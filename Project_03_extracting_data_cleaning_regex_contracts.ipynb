### 1. Social Number (ΑΦΜ/AFM)

Get the social number, or the name of the company from the conclusion of each pdf

One pdf can have more than one social numbers because it can have more than one contract recipients.
In this case you need:
- create a function that searches for all the social numbers ("findall")
- test it
- create a separate df with a new column that will apply the function to each row and if a pdf has more than one social numbers it will add it as a new row (".explode")
- merge your new df with your old df using the column doc_name

```python
import re

def get_afms(text):
    primary_pattern = r"Α\.?Φ\.?Μ\.?.*?(\d+)"
    secondary_pattern = r"(\b\d{9}\b)"
    
    matches = re.findall(primary_pattern, text, flags=re.DOTALL)
    if not matches:
        matches = re.findall(secondary_pattern, text, flags=re.DOTALL)
    
    return matches
```

#### re.DOTALL

The re.DOTALL flag in Python's re module is used when working with regular expressions.
This flag changes the behavior of the dot (.) metacharacter so that it matches any character, including newline characters.
Without this flag, the dot would match any character except a newline.
And since after a lot of trial and error we saw that social numbers can have a new line character before the number starts.
in this case this ([.\n]*?) didn't work

```python
#check the first one

get_afms(df_pdfs["text"][0])
```

```python
df_pdfs['text'] = df_pdfs['text'].fillna('').astype(str)

df_social_number = (
    df_pdfs
    .assign(
        social_number_recipient = lambda df: df['text'].apply(get_afms),
    )
    [[
        "doc_name",
        "social_number_recipient",
    ]]
    .explode("social_number_recipient")
)
df_social_number
```

```python
df_afm = pd.merge(df_pdfs, df_social_number, on="doc_name", how="left")
df_afm
```

### 2. Get more info from the minirsty's decision at the end of each contract to get the recipient

In each document of a contract there is a conclusion that always starts with the word "ΑΠΟΦΑΣΙΖΟΥΜΕ" meaning "we have decided" and usually in the first couple of paragraphs there is more information regarding both the contract and the recipient. This can be helpful for cross-checking the name of the companies and the social numbers, but also for getting information for the recipients without a social number. 

```python
def get_decision(text):
    patterns = [
        r"ΑΠΟΦΑΣΙΖΟΥΜΕ(.*)",
        r"Α Π Ο Φ Α Σ Ι Ζ Ο Υ Μ Ε(.*)",
        r"Αποφασίζουμε(.*)",
        r"Α π ο φ α σ ί ζ ο υ μ ε(.*)",
        r"Α Π Ο Φ Α Σ Ι Ζ Ο Υ ΜΕ(.*)",
        r"ΣΥΜΒΑΣΗ ΑΝΑΘΕΣΗΣ(.*)",
        r"Α Π Ο Φ Α ΣΙ Ζ Ο Υ Μ Ε(.*)",
        r"ΑΝΑΘΕΤΕΙ(.*)"
    ]
    
    for pattern in patterns:
        matches = re.findall(pattern, text, flags=re.DOTALL)
        if matches:
            return matches[0]
    
    return None  
```

```python
df_conclusion= (
    df_unique
    .assign(
        conclusion = lambda df: df['text'].apply(get_decision),
    )
)
df_conclusion
```

```python
def get_recipient(text):
    if text is None:
        return None
        
    patterns = [
        r"επιχείρησης\s*?([\w\s,]+)\s\d{9}\b",     # Capture after 'επιχείρησης' before a 9-digit number
        r"εταιρείας\s*?([\w\s,]+)\s\d{9}\b",       # Capture after 'εταιρείας' before a 9-digit number
        r"επιχείρησης.*?«([^»]*)»",               # Capture inside Greek quotes after 'επιχείρησης'
        r"«([^»]*)»",                             # Capture inside any Greek quotes
        r"(.{0,100}?)\b\d{9}\b",                  # Capture up to 100 characters before a 9-digit number
        r"εταιρεία\s*?([\w\s,]+)\s\d{9}\b",        # Capture after 'εταιρεία' before a 9-digit number
        r"εταιρεία\s*?([\w\s,]+)\sμε",  
        r"“([^”]*)”",                             # Capture inside any Greek quotes
        r"[Α-ΩA-Z][Α-ΩA-Z\-\,\.\s\&«»""]{4,}"
    ]
    
    for pattern in patterns:
        matches = re.findall(pattern, text, flags=re.DOTALL)
        if matches:
            return matches[0]
    
    return None
```

```python
recipient= (
    df_conclusion
    .assign(
        recipient = lambda df: df['conclusion'].apply(get_recipient),
    )
)
recipient
```

```python
def get_payment_amount(text):
    if text is None:
        return None
    
    patterns = [
        # r"\d{1,3}(?:\.\d{3})*,\d{2}\s€",              # Matches "1.234,56 €"
        # r"\(\d{1,3}(?:\.\d{3})*,\d{2}€\)",            # Matches "(1.234,56€)"
        # r"\(€ \d{1,3}(?:\.\d{3})*,\d{2}\)",           # Matches "(€ 1.234,56)"
        # r"\(€\s*\d{1,3}(?:\.\d{3})*,\d{2}\)",         # Matches "(€ 1.234,56)"
        # r"\d{1,3}(?:\.\d{3})*,\d{2} ευρώ",            # Matches "1.234,56 ευρώ"
        # r"\d+ ευρώ",                                  # Matches "1234 ευρώ"
        # r"\(\d+€\)",                                  # Matches "(1234€)"
        # r"#\d{1,3}(?:\.\d{3})*,\d{2}€#",              # Matches "#1.234,56€#"
        # r"#\s\d{1,3}(?:\.\d{3})*,\d{2}#",             # Matches "# 1.234,56#"
        # r"#\d{1,3}(?:\.\d{3})*,\d{2}#",               # Matches "#1.234,56#"
        # r"ευρώ\s\d{1,3}(?:\.\d{3})*,\d{2}€",          # Matches "ευρώ 1.234,56€"
        # r"ευρώ\s\d{1,3}(?:\.\d{3})*,\d{2}\s*\(.*?\)", # Matches "ευρώ 1.234,56€ (συμπεριλαμβανομένου)"
        # r"\d{1,3}(?:\.\d{3})*,\d{2}\sΕΥΡΩ",           # Matches "4996,95 ΕΥΡΩ",
        r"\d[\d\,\.]*\s*(?:€|ευρώ)"
    ]
        

    for pattern in patterns:
        matches = re.findall(pattern, text, flags=re.DOTALL)
        if matches:
            return matches[0]
    
    return None
```

```python
payments= (
    recipient
    .assign(
        payments = lambda df: df['conclusion'].apply(get_payment_amount),
    )
)
payments
```

```python
payments.to_csv("mit_contracts_afm_payments_recipient.csv")
```

### 4. Direct contracts

To search for direct contracts ("απευθείας ανάθεση") we can use this (?i) in our regex to ensure that the regex search is case-insensitive, matching both lowercase and uppercase versions of the phrase ("απευθείας ανάθεση" or "Απευθείας ανάθεση")

```python
direct_contracts = (
    payments
    .assign(
        direct_emergency_contracts=lambda df: df['subject'].apply(
            lambda x: 'direct' if pd.notna(x) and re.search(r"(?i)απευθείας ανάθεση", x) else 'no'))
)
direct_contracts
```

You can check how many they are

```python
direct_contracts["direct_emergency_contracts"].value_counts()
```

## Government - recipients connection

The issue_date column can be used to identify the political party/ministers in power when the payments were conducted, to check for connections between recipient and government figures

- create an empty column
- define the start and end of the previous government (SYRIZA) and the start of the current government (New Democracy)
- check if you have payments from before the previous government 

```python
direct_contracts['government_party'] = None

start_syriza_gov = pd.Timestamp('2015-09-23')
end_syriza_gov = pd.Timestamp('2019-12-08')
start_nd_gov = pd.Timestamp('2019-12-09')

direct_contracts.loc[
    (direct_contracts['issue_date_converted'] >= start_syriza_gov) & 
    (direct_contracts['issue_date_converted'] <= end_syriza_gov), 'government_party'
] = 'SYRIZA'

direct_contracts.loc[
    direct_contracts['issue_date_converted'] >= start_nd_gov, 'government_party'
] = 'New Democracy'
```
